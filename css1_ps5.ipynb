# 1: Creating functions

```
# Create lists
test_list_1 = [8, 8, 3, 5, 5, 8, 1, 4, 8, 6, 3, 10, 9]
test_list_2 = [8, 12, 3, 5, 5, 8, 1, 4, 8, 3, 10, 9]

# Function for mean of a list

def mean_func(x):
  """
  purpose: find the mean of a list
  
  inputs:
  x = list of numbers the mean is for
  """

  mean_num = round(sum(x)/len(x), 2)
  return mean_num

mean1 = mean_func(test_list_1)
print (mean1)

mean2 = mean_func(test_list_2)
print (mean2)
```

```
# Function for median of a list

def median_func(x):
  """
  purpose: find the median of a list
  
  inputs:
  x = list of numbers the median is for
  """

  x.sort()

  if len(x) % 2:
    median_num = x[len(x) // 2]
  else:
    median_num = ((x[len(x) // 2]) + (x[len(x) // 2 - 1])) / 2
  return median_num

median1 = median_func(test_list_1)
print (median1)

median2 = median_func(test_list_2)
print (median2)
```

```
# Function for variance of elements in a list

def var_func(x):
  """
  purpose: find the variance of each element in a list
  
  inputs:
  x = list of numbers to find the variance of
  """

  mean_num = round(sum(x)/len(x), 2)

  var_num = [round(((num - mean_num)**2), 2) for num in x]
  return var_num

var1 = var_func(test_list_1)
print (var1)

var2 = var_func(test_list_2)
print (var2)
```

```
# Function to magnify values by 10

def mag10_func(x):
  """
  purpose: magnify each value in a list by 10
  
  inputs:
  x = list of numbers to magnify by 10
  """

  x.sort()

  mag10_num = [(num*10) for num in x]
  return mag10_num

mag10_1 = mag10_func(test_list_1)
print (mag10_1)

mag10_2 = mag10_func(test_list_2)
print (mag10_2)
```

# 2: Basic Calculator

```
# Defining functions

# Addition function
def add_nums(num1, num2):
  """
  purpose: a function that adds numbers

  inputs:
  num1 = the first number
  num2 = the second number
  """
  
  a = num1 + num2
  return a

# Subtraction function
def subtract_nums(num1, num2):
  """
  purpose: a function that subtracts numbers

  inputs:
  num1 = the first number
  num2 = the second number
  """

  b = num1 - num2
  return b

# Multiplication function
def multiply_nums(num1, num2): 
  """
  purpose: a function that multiples numbers

  inputs:
  num1 = the first number
  num2 = the second number
  """

  c = num1 * num2
  return c

# Division
def divide_nums(num1, num2):
  """
  purpose: a function that divides numbers

  inputs:
  num1 = the first number
  num2 = the second number
  """

  try:
    d = num1 / num2
    return d

  except ZeroDivisionError:
    print ('Error: cannot divide by 0')
```

```
# Loop for performing the calculations

for i in range(0,5):

# Input for num1

  while True:
    
    num1 = input("Enter first number: ")
  
    if num1.isdigit():
      num1 = int(num1)
      break
    else:
      print ('Error: Please enter an integer')

# Input for num2

  while True:
    
    num2 = input("Enter second number: ")

    if num2.isdigit():
      num2 = int(num2)
      break
    else:
      print ('Error: Please enter an integer')

# Input for operation

  while True:

    operation = input("Please enter '+' for addition, '-' for subtraction, '*' for multiplication, or '/' for division: ")

    if operation == '+':
      print ('Sum = ', add_nums(num1, num2))
      break
    elif operation == '-':
      print ('Difference = ', subtract_nums(num1, num2))
      break
    elif operation == '*':
      print ('Product = ', multiply_nums(num1, num2))
      break
    elif operation == '/':
      print ('Quotient = ', divide_nums(num1, num2)) 
      break
    else:
      print ('Error: Please enter one of the given operation symbols')
```

```
# Write function

def accept_login(users, username, password):
  """
  purpose:
  inputs:
  """
  
  if username in users and password == users[username]:
    return True
  else:
    return False

users = {"user1": "password1", "user2": "password2", "user3": "password3"}

# check if username-password pairs are accepted

check_users = {'wronguser': 'wrongpassword', 'user1': 'password1', 'user2': 'password1', 'user3': 'password2'}

for key,value in check_users.items():
  if accept_login(users, key, value):
    print ('Login successful')
  else:
    print ('Login failed...')
```
