```
# square numbers in the file

b_list = []

with open('CSS1_week6.txt', 'r') as f:

  for line in f:
    line = line.split()
    sub_list = []
    for numbers in line:
      numbers = int(numbers)
      sqr_lists = numbers ** 2
      sub_list.append(sqr_lists)

    b_list.append(sub_list)

print (b_list)
```

```
# write text file with each list of b_list as a separate line
with open('CSS_week6_new.txt', 'w') as f:
  for i in b_list:
    for num in i:
      f.write('{0:<3}'.format(num))
    f.write('\n')

%pycat CSS_week6_new.txt
```

```
%%writefile new_math_tools.py

def sum_of_anything(*numbers):
  x = 0
  for i in numbers:
    x = x + i

  return x

def concatenate_everything(**strings):
  y = ''
  for i in strings.values():
    y = y + i
  
  return y
```
