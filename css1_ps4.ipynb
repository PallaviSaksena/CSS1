# 1: Key-values search

```
# Create dictionary of NFL teams

d = {'SF': '49ers', 'Pittsburgh': 'Steelers', 'Seattle': 'Seahawks', 'New York': 'Giants', 'Arizona': 'Cardinals'}
print (d)

print () # for spacing

# Create list of all the values (nicknames)

nicknames = []

for key, value in d.items():
  nicknames = list(d.values())

print (nicknames)

print () # for spacing

# Create list of all the keys (states)

states = []

for key, value in d.items():
  states = list(d.keys())

print (states)

print () # for spacing

# Check if states are in dictionary

keys_to_search = ['Arizona', 'Chicago', 'Tennessee', 'New York', 'Atlanta']

for state in keys_to_search:
  if state in d:
    print (str(state), "is one of the keys with nickname", d.get(state))
  else:
    print (str(state), "is not one of the keys")
```

# 2: List comprehension operations

```
# Create list of filenames

filenames = ['f1.png', 'f2.txt', 'f3.txt', 'f4', 'f5.png', 'f6.png', 'f7', 'f8.txt']

# Create list of filenames that have .png

png_files = [File for File in filenames if '.png' in File]
print (png_files)

print () # for spacing

# Create list of valid filenames (.png or .txt)

valid_files = [File for File in filenames if '.png' in str(File) or '.txt' in str(File)]
print (valid_files)

print () # for spacing

# Create list of invalid filenames

invalid_files = [File for File in filenames if '.png' not in File and '.txt' not in File]
print (invalid_files)

print () # for spacing

# Add .png extension to invalid files

add_extension = [File + '.png' for File in invalid_files if '.png' not in File or '.txt' not in File]
print (add_extension)
```

# 4: Dictionary key-values, pop, update

```
# Create list of tuples

list_of_tuples = [('Jeff', 'tennis'), ('Jenna', 'soccer'), ('Prisha', 'hockey'), ('Lee', 'gymnastics'), ('Ryan', 'swimming')]

# Create dictionary from list of tuples

dict1 = dict(list_of_tuples)
print (dict1)

print () # for spacing

# Pop out last key-value pair

(item_key, item_value) = dict1.popitem()
print (item_key, item_value)

print () # for spacing

print (dict1)

print () # for spacing

# Access popped name from dictionary

print (dict1.get(item_key))

print () # for spacing

# Create second dictionary

dict2 = {'Bobby': 'chess', 'Sachin': 'cricket'}

# Update first dictionary with contents of second dictionary

dict1.update(dict2)

print (dict1)

print () # for spacing

# Access values from updated dictionary and convert to list

list_dict_values = list(dict1.values())

# Print last element of list

print (list_dict_values[-1])

print () # for spacing

# Acess all keys from updated dictionary and convert to list

list_dict_keys = list(dict1.keys())

# Print all elements that have first letter B

B_elements = [word for word in list_dict_keys if "B" in str(word)]
print (B_elements)
```
