```
# load file as df
df = pd.read_csv('GME.csv')
df.head()
```

```
# make dataframe df1 with only NON-NA values of 'adj close' column

df1 = df[df['Adj Close'].notna()]

display(df1)
```

```
# make dataframe df2 which fills in all NA values in 'adj close' column with average 'adj close' value

df2 = df.fillna(df['Adj Close'].mean())

display(df2)
```

```
# report amount of rows in df1 and df2

print ('# rows in df1:', len(df1))
print ('# rows in df2:', len(df2))
```

```
# calculate means of df1 and df2

df1_mean = round(df1['Adj Close'].mean(), 2)
df2_mean = round(df2['Adj Close'].mean(), 2)

print ('df1 mean:', df1_mean)
print ('df2 mean:', df2_mean)

if df1_mean == df2_mean:
  print ("True")
else:
  print ('False')
```

```
# make dataframe df3 that includes only rows with 'adj close' values higher than average 'adj close' values

df3 = df[(df['Adj Close'] > df1_mean)]
display (df3)
```

```
# calculate std deviation of df3 and df1

df1_std = round(df1['Adj Close'].std(), 2)
df3_std = round(df3['Adj Close'].std(), 2)

print ('standard deviation of df1:', df1_std)
print ('standard deviation of df3:', df3_std)
```

```
# store file as df

df = pd.read_csv('CSS1_PS8_Q2.csv')
display (df.head())
```

```
# PART 1 - cleaning and setting dataframe

# delete day 2021-01-01T from Time column

df['Time'] = df['Time'].str.replace('2021-01-01T', '')
df.head()
```

```
# make first column of dataframe as row index

df = df.set_index('Time')
df.head()
```

```
# isolate just the rows (indices) where Mean is NaN

print (df[df.isna()])
```

```
# delete these rows from dataframe

df = df.dropna()
df.head()
```

```
# PART 2 - basic operations

# calculate mean dry bulb temperature

print ('Mean dry bulb temperature:', df['HourlyDryBulbTemperature'].mean())
```

```
# find and print row with minimum hourly relative humidity

print ('Min hourly relative humidity:', df['HourlyRelativeHumidity'].min())
print ()

pd.set_option('display.max_columns', None)
print (df[df['HourlyRelativeHumidity'] == df['HourlyRelativeHumidity'].min()])
```

```
# find and print all rows with hourly visibility as 10

print (df[(df['HourlyVisibility'] == 10)])
```
