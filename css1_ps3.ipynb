# 1: Tuple unpacking

```
# Create tuple

my_tuple = ('Pallavi', 'Psychology', 2024)
print (my_tuple)

# Unpack tuple

(Name, Major, Class_year) = my_tuple

print ('Name:', Name)
print ('Major:', Major)
print ('Class year:', Class_year)

# Reassign major variable

my_tuple = ('Pallavi', 'CSS', 2024)
new_major = my_tuple[1]

print ('Name:', Name)
print ('Major:', new_major)
print ('Class year:', Class_year)

# If...else statement to check if the majors are the same

if new_major == Major:
  print ("Yes they are the same!")
else:
  print ("No they're not the same!")
```

# List concatenating and indexing

```
# Create lists

fruits = ["apple", "mango", "banana"]
veggies = ["carrot", "cabbage", "potato"]
flowers = ["rose", "lotus", "sunflower"]

# print list name for a given word

word = "potato"

if word in fruits:
  print ("fruit")
elif word in veggies:
  print ("veggie")
elif word in flowers:
  print ("flower")
else:
  print ("not found")

word = "sunflower"

if word in fruits:
  print ("fruit")
elif word in veggies:
  print ("veggie")
elif word in flowers:
  print ("flower")
else:
  print ("not found")

word = "mango"

if word in fruits:
  print ("fruit")
elif word in veggies:
  print ("veggie")
elif word in flowers:
  print ("flower")
else:
  print ("not found")

word = "avocado"

if word in fruits:
  print ("fruit")
elif word in veggies:
  print ("veggie")
elif word in flowers:
  print ("flower")
else:
  print ("not found")

# Concatenate lists

final_list = [fruits, veggies, flowers]
print (final_list)

# Index into list of lists

print (final_list[2][0])

print (final_list[1][1])
```

# 3: Print patterns using for loops

```
# Print * pattern

for i in range(0, 5):
  for j in range(0, i + 1):
    print ("*", end = " ")
  print (" ")

# Print number pattern 1-5

for i in range(5, 0, -1):
  for j in range(1, i + 1):
    print (j, end = " ")
  print (" ")

# Print number pattern 5-25

for i in range (0, 5):
  for j in range (1, i + 2):
    print (j * 5, end = " ")
  print (" ")
```

# 4: For loops and if else statements

```
# Print all even numbers from 2-50

x = list(range(2, 51))
even_num = []

for i in x:
  if i % 2 == 0:
    even_num.append(i)

print (even_num)

# Print sum of all odd numbers from 50-100

y = list(range(50, 101))

odd_num = []

for i in y:
  if i % 2 == 1:
    odd_num.append(i)

Sum = sum(odd_num)
print (Sum)

# For Loop and If...Else to square and cube numbers

List_Q4c = [2, 4, 6, 7, 10, 15, 23, 25]

List_Q4c_mod = List_Q4c.copy()

for i in range(len(List_Q4c_mod)):
  if List_Q4c_mod[i] % 5 == 0:
    List_Q4c_mod[i] = List_Q4c_mod[i]**2
  else:
    List_Q4c_mod[i] = List_Q4c_mod[i]**3

print (List_Q4c)
print (List_Q4c_mod)
```

# 5: Voting age eligibility determiner

```
# Create a list of ages

ages = [10, 16, 18, 22, 35, 13]

# Print ages with a for loop

for age in ages:
  print (age)

# Use an if else statement

for age in ages:
  if (age > 18):
    print ("You are eligible to vote!")
  if (age == 18):
    print ("You are eligible to vote this year. You'll be voting for the first time!")
  if (age < 18):
    x = 18 - age
    print ("You can't vote this year. You have to wait " + str(x) + " years to vote!")
```
