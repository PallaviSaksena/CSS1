# 1: Fill array with 5000 random numbers

```
# Traditional approach - for loop

# import numpy
import numpy as np

# initialize empty list
rand_nums = []

# loop 5000 times and append random number each time + time it
%timeit for i in range(0, 5000): rand_nums.append(np.random.random())
```

```
# For loop with numpy method

# import numpy
import numpy as np

# pre-allocate numpy array of zeroes
arr = np.zeros(5000)

# loop 5000 times and assign random number + time it
%timeit for i in np.arange(5000): arr[i] = np.random.random()
```
